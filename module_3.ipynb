### 7.8 Create a list called surprise with the elements "Groucho", "Chico", and "Harpo".

```python
surprise = ["Groucho", "Chico", "Harpo"]
print(surprise)
```

### 7.9 Lowercase the last element of surprise, reverse it, and then capitalize it

```python
last = surprise[-1].lower()   # "harpo"
last = last[::-1]             # "oprah"
last = last.capitalize()      # "Oprah"
print(last)
```

### 9.2  Define a generator function called get_odds() that returns the odd numbers from range(10). 

```python
def get_odds():
    for num in range(10):
        if num % 2 == 1:
            yield num

count = 0
for value in get_odds():
    count += 1
    if count == 3:
        print(value)
        break

```
